wap to find the net salary of employee using inheritance, create 3 classes employee, perks and net_salary
    
    make employee class as abstract class, employee class have 4 methods:
    1. getemployee() --> takes user input for id,name and basic salary
    2. employeedetails() --> print id,name and basic salary
    3. getsalary() --> return a basic salary
    4. emp_id() --> abstract method

    class perks(employee)
    1. calc_perks() --> calculate DA, HRA, PF
            DA - 35% of Basic salary
            HRA - 17% of Basic salary
            PF - 12% of Basic salary
    2. print_perks() --> print DA, HRA, PF
    3. total_perks() --> DA + HRA - PF
    4. emp_id()
    
    class net_salary(perks)
    1. get_total() --> calculate total salary

```python
from abc import ABC,abstractmethod

class employee(ABC): #----------------> ABSTRACT CLASS
    def __init__(self):
        self.getemployee()
        self.employeedetails()
        
    def getemployee(self):
        self.id=input("enter id: ")
        self.name=input("enter name: ")
        self.bs=int(input("enter basic salary: "))
        
    def employeedetails(self):
        print("id: ",self.id)
        print("name: ",self.name)
        print("basic salary: ",self.bs)
        
    def getsalary(self):
        return self.bs
    
    @abstractmethod
    def emp_id(self):
        pass
    
class perks(employee):
    def calc_perks(self):
        self.DA=0.35*self.bs
        self.HRA=0.17*self.bs
        self.PF=0.12*self.bs
        
    def print_perks(self):
        print("DA: ",self.DA)
        print("HRA: ",self.HRA)
        print("PF: ",self.PF)
        
    def total_perks(self):
        self.tp=self.DA+self.HRA-self.PF
      
    def emp_id(self):
        pass
    
class net_salary(perks):
    def __init__(self):
        self.getemployee()
        self.employeedetails()
        self.calc_perks()
        self.print_perks()
        self.total_perks()
        self.get_total()
        
    def get_total(self):
        self.ts=self.bs+self.tp
        print(self.ts)
        
    def emp_id(self):
        pass
    
e1=net_salary()
```
