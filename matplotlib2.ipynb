```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([76,50,76,69])
mylabel=['Water','Banana','Kiwi','Apple']
wadge=[0.1,0,1,0]
mycolors=['yellow','skyblue','pink','violet']
plt.pie(x,labels=mylabel,explode=wadge,shadow=True,colors=mycolors)
plt.legend(title='four fruits')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
x=np.array([76,50,76,69])
mylabel=['Water','Banana','Kiwi','Apple']
wadge=[0.1,0,0,0]
mycolors=['yellow','skyblue','magenta','violet']
plt.pie(x,labels=mylabel,explode=wadge,shadow=True,colors=mycolors,autopct="%1.2f%%")
plt.legend(loc='lower right')
plt.show()
```
